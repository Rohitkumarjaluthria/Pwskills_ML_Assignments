Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import matplotlib.pyplot as plt

plt.hist(flight_data['Price'], bins=20, edgecolor='black')

plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

price_range = max_price - min_price

print("Price Range: {} - {}".format(min_price, max_price))
print("Range of Prices: {}".format(price_range))
print("Minumum Price = ", min_price)
print("Maximum Price = ", max_price)
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot([flight_data[flight_data['Airline'] == airline]['Price'] for airline in flight_data['Airline'].unique()],
            labels=flight_data['Airline'].unique())

plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')

plt.xticks(rotation=90)
plt.show()
```

Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does
the dataset have?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('googleplaystore.csv')

# Get the dimensions of the dataset
rows, columns = df.shape

# Print the number of rows and columns
print("Number of rows:", rows)
print("Number of columns:", columns)
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

category_ratings = df.groupby('Category')['Rating'].median().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.boxplot(category_ratings, vert=False)
plt.title('Ratings by App Category')
plt.xlabel('Rating')
plt.ylabel('App Category')
plt.show()
```

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
missing_values = df.isnull().sum()
print(missing_values)

missing_percentage = (missing_values / len(df)) * 100
print(missing_percentage)
```

Missing values can affect the accuracy and reliability of your results. They may lead to biased insights or inaccurate conclusions if not handled properly. The impact of missing values depends on the specific analysis or task you are performing. For example:

If you are analyzing the average rating of apps by category and there are missing values in the "Rating" column, it may affect the accuracy of the average ratings for certain categories.

If you are investigating the correlation between app size and installs and there are missing values in the "Size" or "Installs" columns, it may affect the validity of your correlation analysis.

Missing values in columns such as "Reviews" or "Price" may impact statistical analysis or machine learning models that rely on those features.

Depending on the impact and the specific analysis you are conducting, you have several options for handling missing values, such as:

Removing the rows or columns with missing values if they are not significant in your analysis.

Imputing the missing values using techniques like mean imputation, median imputation, or regression imputation.

Using advanced imputation methods like multiple imputation or K-nearest neighbors imputation.

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import matplotlib.pyplot as plt

# Calculate average prices by app type
avg_prices = df.groupby('Type')['Price'].mean()

# Plotting the bar chart
avg_prices.plot(kind='bar')

# Adding labels and title
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv('googleplaystore.csv')

# Create a frequency table of app names and their install counts
app_frequency = df[['App', 'Installs']].groupby('App').sum().sort_values(by='Installs', ascending=False)

# Display the top 10 most popular apps
top_10_apps = app_frequency.head(10)
print(top_10_apps)
```
